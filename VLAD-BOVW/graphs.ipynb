```python
import matplotlib.pyplot as plt
```

```python
data = {
    "kmeans-k": [8, 10, 15, 50],
    "SIFT": [0.3112518974, 0.3076348107, 0.3039006521, 0.2920154617],
    "SURF": [0.3055166667, 0.3078666667, 0.3094166667, 0.3028333333]
}
```

```python
fig, ax = plt.subplots()
ax.plot(data["kmeans-k"], data["SIFT"], 'r^-', label="SIFT")
ax.plot(data["kmeans-k"], data["SURF"], 'g^-', label="SURF")
plt.xlabel("k means - Number of clusters")
plt.ylabel("Accuracy")

legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')


plt.show()
```
